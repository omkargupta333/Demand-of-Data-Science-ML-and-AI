```python
# Importing relevant libraries
import pandas as pd
from pytrends.request import TrendReq   # connect to google
import matplotlib.pyplot as plt
trends = TrendReq() 
```

# Data Science

```python
# Load Data
trends.build_payload(kw_list=["Data Science"])
data = trends.interest_by_region()
data = data.sort_values(by="Data Science", ascending=False)
data = data.head(20)
data
```

```python
# Create plot
plt.figure(figsize=(10, 5))
plt.bar(data.index, data["Data Science"])
plt.title("Interest in Data Science by Region")
plt.xlabel("Region")
plt.ylabel("Interest")
plt.xticks(rotation=75)

# Enhance plot background
plt.grid(axis="y", alpha=0.5)
plt.axhline(y=0, color="gray", linewidth=5)
plt.gca().spines["right"].set_visible(False)
plt.gca().spines["top"].set_visible(False)

plt.show()
```

```python
data = TrendReq(hl='en-US', tz=360)
data.build_payload(kw_list=['Data Science'])
data = data.interest_over_time()
fig, ax = plt.subplots(figsize=(15, 12))
data['Data Science'].plot()
plt.style.use('fivethirtyeight')
plt.title('Total Google Searches for Data Science', 
          fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Total')
```
